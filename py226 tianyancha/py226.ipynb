```python
class TycAPI:
    def __init__ (this):
        this.count = 0
    
    def __str__ (this):
        return f'[TycAPI(use api count#{this.count})]'
    
    def __repr__ (this): return str(this)

    def api (this, page, file, folder, **kw):
        if not kw or not [v for v in kw.values() if v is not None]: return page   
        page = page + '?' + '&'.join([f'{k}={kw[k]}' for k in kw if kw[k] is not None])
        rst = requests.get(page, headers=headers)
        jrst = rst.json()
        if not jrst.get('error_code', 0) or jrst.get('error_code', 0) == 300000:
            this.count += 1
            if (savePath): fprint (rst.text, file=f'{file}.json', path=os.path.join(savePath,folder))
        return {} if jrst.get('error_code', 0) == 300000 else jrst.get('result', jrst)
    
    def search (this, keyword):
        page = 'http://open.api.tianyancha.com/services/open/search/2.0'
        folder = 'search'
        return this.api(page, keyword, folder, word=keyword)
    
    def trademarkSearch (this, keyword, pageNum=1):
        page = 'http://open.api.tianyancha.com/services/open/ipr/tm/search'
        folder = 'trademarkSearch'
        return this.api(page, f'{keyword}#{pageNum}', folder, keyword=keyword, pageNum=pageNum)
    
    def patentSearch (this, keyword, pageNum=1):
        page = 'http://open.api.tianyancha.com/services/open/ipr/patents/search'
        folder = 'patentSearch'
        return this.api(page, f'{keyword}#{pageNum}', folder, keyword=keyword, pageNum=pageNum)
    
    def baseInfo (this, keyword):
        page = 'http://open.api.tianyancha.com/services/open/ic/baseinfoV3/2.0'
        folder = 'baseInfo'
        return this.api(page, keyword, folder, keyword=keyword)
    
    def teamMember (this, keyword, pageNum=1):
        page = 'http://open.api.tianyancha.com/services/open/cd/findTeamMember/2.0'
        folder = 'teamMember'
        return this.api(page, f'{keyword}#{pageNum}', folder, keyword=keyword, pageNum=pageNum)

    def product (this, keyword, pageNum=1):
        page = 'http://open.api.tianyancha.com/services/open/cd/getProductInfo/2.0'
        folder = 'product'
        return this.api(page, f'{keyword}#{pageNum}', folder, keyword=keyword, pageNum=pageNum)
    
    
    def trademark (this, keyword, pageNum=1):
        page = 'http://open.api.tianyancha.com/services/open/ipr/tm/2.0'
        folder = 'trademark'
        return this.api(page, f'{keyword}#{pageNum}', folder, keyword=keyword, pageNum=pageNum)
    
    def patent (this, keyword, pageNum=1):
        page = 'http://open.api.tianyancha.com/services/open/ipr/patents/2.0'
        folder = 'patent'
        return this.api(page, f'{keyword}#{pageNum}', folder, keyword=keyword, pageNum=pageNum)

    def softwareCopyright (this, keyword, pageNum=1):
        page = 'http://open.api.tianyancha.com/services/open/ipr/copyReg/2.0'
        folder = 'softwareCopyright'
        return this.api(page, f'{keyword}#{pageNum}', folder, keyword=keyword, pageNum=pageNum)
    
    def website (this, keyword, pageNum=1):
        page = 'http://open.api.tianyancha.com/services/open/ipr/icp/3.0'
        folder = 'website'
        return this.api(page, f'{keyword}#{pageNum}', folder, keyword=keyword, pageNum=pageNum)
```

## mysql插入定义

```python
db = pymysql.connect(mysqlIp, mysqlUser, mysqlPassword,port=mysqlPort,charset='utf8',db=mysqlDb)
cursor = db.cursor()
addCursor(cursor)
```

```python
updateGap = 2592000
```

```python
mysqlDb = 'tianyancha'
def __std (s):
    if s is None or s == '': return 'null'
    if type(s) in [list, dict, set]: s = str(s)
    return repr(s)

def checkEnterprise (id):
    maxUpdateTime = str(ts(getNow()) - td(f'{updateGap}s'))
    rst = select('id', f'`{mysqlDb}`.`enterprise`', where=f'id={__std(id)} and updated_at>{__std(maxUpdateTime)}')
    return bool(len(rst))

def checkKeyword (keyword):
    maxUpdateTime = str(ts(getNow()) - td(f'{updateGap}s'))
    rst = select('*', f'`{mysqlDb}`.`keyword`', where=f'keyword={__std(keyword)} and updated_at>{__std(maxUpdateTime)}')
    return bool(len(rst))

def insert (table, *args, ignore=''):
    sqlInsert = f'''insert {'ignore' if ignore else ''} into {table} values %s;'''
    value = '(' + ','.join([__std(x) for x in args]) + ')'
    sql(sqlInsert % value)

def insertMulti (table, values, ignore=''):
    if len(values) == 0: return
    sqlInsert = f'''insert {'ignore' if ignore else ''} into {table} values %s;'''
    values = ['(' + ','.join([__std(y) for y in x]) + ')' for x in values]
    values = ','.join(values)
    sql(sqlInsert % values)

def delete (table, col, eid):
    sqlDelete = f'''delete from {table} where {col}={eid};'''
    sql(sqlDelete)


def insertEnterprise (**kw):
    '''
    (id, regNumber, creditCode, taxNumber, name, name_en, 
    alias, historyNames, historyNameList, legalPersonName,
    regCapital, industry, regLocation, businessScope, phoneNumber, 
    websiteList, email, staffNumRange, nStaff, nTeamMember, nProduct, 
    nTrademark, nPatent, nSoftwareCopyright, nWebsite)
    '''
    now = getNow()
    table = f'`{mysqlDb}`.`enterprise`'
    
    t = select('created_at', table, where=f'id={__std(id)}')
    createdAt = t['created_at'][0] if len(t) else now
    
    delete (table, 'id', eid)
    cols = ['id', 'regNumber', 'creditCode', 'taxNumber', 'name', 'name_en', 
        'alias', 'historyNames', 'historyNameList', 'legalPersonName',
        'regCapital', 'industry', 'regLocation', 'businessScope', 'phoneNumber', 
        'websiteList', 'email', 'staffNumRange', 'nStaff', 'nTeamMember', 'nProduct', 
        'nTrademark', 'nPatent', 'nSoftwareCopyright', 'nWebsite']
    args = [kw.get(c, None) for c in cols]
    insert(table, *args, createdAt, now)

def insertKeyword (keyword):
    now = getNow()
    table = f'`{mysqlDb}`.`keyword`'
    
    t = select('created_at', table, where=f'keyword={__std(keyword)}')
    createdAt = t['created_at'][0] if len(t) else now
    
    delete (table, 'keyword', keyword)
    insert(table, keyword, createdAt, now)


def insertStaffs (eid, staffs):
    '''
    (sid, name, type, typeJoin)
    '''
    table = f'`{mysqlDb}`.`staff`'
    delete(table, 'eid', eid)
    insertMulti(table, [
        (eid, *staff) for staff in staffs
    ], ignore='ignore')

def insertHuman (**human):
    id = human.get('id', None)
    name = human.get('name', None)
    description = human.get('description', None)
    icon = human.get('icon', None)
    
    if id is None: return
    
    insert(f'`{mysqlDb}`.`human`', id, None, None, None, ignore=True)
    if name is not None:
        sql(f'''
            update `{mysqlDb}`.`human` set `name` = {__std(name)} 
            where (`id` = {__std(id)});
            ''')
    if description is not None:
        sql(f'''
            update `{mysqlDb}`.`human` set `description` = {__std(description)} 
            where (`id` = {__std(id)});
            ''')
    if icon is not None:
        sql(f'''
            update `{mysqlDb}`.`human` set `icon` = {__std(icon)} 
            where (`id` = {__std(id)});
            ''')

def insertTeamMembers (eid, teamMembers):
    '''
    (hid, name, isDimision, title, icon, description)
    '''
    table = f'`{mysqlDb}`.`teamMember`'
    delete(table, 'eid', eid)
    insertMulti(table, [
        (eid, *tm) for tm in teamMembers
    ], ignore='ignore')

def insertProducts (eid, products):
    '''
    (hangye, yewu, product, logo, setupTime)
    '''
    table = f'`{mysqlDb}`.`product`'
    delete(table, 'eid', eid)
    insertMulti(table, [
        (eid, *p) for p in products
    ])

def insertTrademarks (eid, trademarks):
    '''
    (id, intCls, title, regNo, applicantCn, pic)
    '''
    table = f'`{mysqlDb}`.`trademark`'
    delete(table, 'eid', eid)
    insertMulti(table, [
        (t[0], eid, *t[1:])
        for t in trademarks
    ], ignore='ignore')

def insertPatents (eid, patents):
    '''
    (id, uuid, title, appNumber, pubNumber, patentNumber, 
    allCatNumber, lawStatus, abstract, 
    applicationTime, pubDate, applicationPublishTime, 
    inventor, applicationNameList, applicationName, agent,
    agency, address, patentType)
    '''
    table = f'`{mysqlDb}`.`patent`'
    delete(table, 'eid', eid)
    insertMulti(table, [
        (p[0], eid, *p[1:]) for p in patents
    ], ignore='ignore')

def insertSoftwareCopyrights (eid, softwareCopyrights):
    '''
    (id, regNum, version, fullName, simpleName, 
    regTime, publishTime, eventTime, authorNationality)
    '''
    table = f'`{mysqlDb}`.`softwareCopyright`'
    delete(table, 'eid', eid)
    insertMulti(table, [
        (sc[0], eid, *sc[1:])
        for sc in softwareCopyrights
    ], ignore='ignore')

def insertWebsites (eid, websites):
    '''
    (webName, ym, website, liscense, companyName)
    '''
    table = f'`{mysqlDb}`.`website`'
    delete(table, 'eid', eid)
    insertMulti(table, [
        (eid, *w)
        for w in websites
    ])
```
